```python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048]
y = [497775, 150760, 50929, 19697, 8520, 3948, 1812, 710, 214, 57, 18, 4]

logx, logy = np.log(x), np.log(y)

p = np.polyfit(logx, logy, 1)
y_fit = np.exp(np.polyval(p, logx))

plt.plot(x, y_fit, 'r--', label='slope = %.1f' % p[0])
plt.plot(x, y)
plt.yscale('log')
plt.xscale('log')
plt.legend();
```

```python
import scipy.stats as stats

def fit_line(x, y):
    logx = np.log(x)
    logy = np.log(y)
    slope, intercept, r_value, p_value, std_err = stats.linregress(logx, logy)
    return slope, intercept, r_value, p_value, std_err

slope, intercept, r_value, p_value, std_err = fit_line(x, y)

y_fit = np.exp(intercept) * np.exp(slope * np.log(x))

# show p value in scientific notation
plt.plot(logx, np.log(y_fit), 'r--', label=f'slope = {slope:.1f} \n p={p_value:.0e}')
plt.plot(logx, logy)
plt.legend();
```
