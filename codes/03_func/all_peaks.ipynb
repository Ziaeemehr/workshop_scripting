```python
import numpy as np
from scipy.signal import argrelextrema
import matplotlib.pyplot as plt

# Generate 2D time series
time = np.linspace(0, 10, 100)
amplitude = np.sin(time) + np.random.rand(100)
series = np.column_stack((time, amplitude))

# plt.plot(series[:, 0], series[:, 1])

# Find local maxima (peaks) of the time series
peaks = argrelextrema(series[:, 1], np.greater)

# Plot the time series and the peaks
plt.plot(series[:, 0], series[:, 1])
plt.plot(series[peaks, 0], series[peaks, 1], 'ro');



# peaks = []
# for i in range(1, len(series)-1):
#     if series[i-1][1] < series[i][1] and series[i][1] > series[i+1][1]:
#         peaks.append(series[i])
# print(peaks)
```
