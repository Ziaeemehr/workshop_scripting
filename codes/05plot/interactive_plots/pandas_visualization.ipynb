### [Pandas Library for Data Visualization in Python](https://stackabuse.com/pandas-library-for-data-visualization-in-python/)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic_data = pd.read_csv("train.csv")
titanic_data.head()
```

```python
titanic_data['Age'].hist();
```

```python
titanic_data['Age'].hist(bins=20);
```

```python
import seaborn as sns
sns.set_style('darkgrid')
titanic_data['Age'].hist(bins=20);
```

```python
titanic_data['Age'].plot(kind='hist', bins=20);
```

```python
titanic_data.plot.line(x='Age', y='Fare', figsize=(5,4));
```

```python
titanic_data.plot.scatter(x='Age', y='Fare', figsize=(5,4));
```

```python
titanic_data.plot.box(figsize=(10,8));
```

```python
titanic_data.plot.hexbin(x='Age', y='Fare', gridsize=30, figsize=(8,6));
```

```python
titanic_data['Age'].plot.kde();
```
